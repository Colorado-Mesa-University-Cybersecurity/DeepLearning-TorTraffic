```python
data = data.dropna()
```

```python
SourcePortList = data[' Source Port']
DestinationPortList = data[' Destination Port']
ProtocolList = data[' Protocol']
FlowDurationList = data[' Flow Duration']
FlowIATMeanList = data[' Flow IAT Mean']
FlowIATStdList = data[' Flow IAT Std']
FlowIATMaxList = data[' Flow IAT Max']
FlowIATMinList = data[' Flow IAT Min']
BwdIATMinList = data[' Bwd IAT Min']
ActiveMeanList = data['Active Mean']
ActiveStdList = data[' Active Std']
ActiveMaxList = data[' Active Max']
ActiveMinList = data[' Active Min']
IdleMeanList = data['Idle Mean']
IdleStdList = data[' Idle Std']
IdleMaxList = data[' Idle Max']
IdleMinList = data[' Idle Min']
```

```python
combinedList = list(zip(SourcePortList, DestinationPortList, ProtocolList, FlowDurationList, FlowIATMeanList, FlowIATStdList, FlowIATMaxList, FlowIATMinList, BwdIATMinList, ActiveMeanList, ActiveStdList, ActiveMaxList, ActiveMinList, IdleMeanList, IdleStdList, IdleMaxList, IdleMinList))
```

```python
combinedList
```

```python
data['label'] = data['label'].map(lambda val : 0 if val == 'nonTor' else 1)
```

```python
data['label']
```

```python
data
```

```python
target_value = list()
```

```python
for d in data['label']:
    target_value.append(d)
```

```python
# target_value
```

```python
from sklearn.model_selection import train_test_split
```

```python
random_state = 1
X = combinedList
y = target_value
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=random_state, stratify=y)
```

```python
print("Number of training samples: {}".format(len(X_train)))
print("Number of testing samples: {}".format(len(X_test)))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
randomForest = RandomForestClassifier(max_depth=2, random_state=random_state)
decisionTree = DecisionTreeClassifier(random_state = random_state)
```

```python
randomForest.fit(X_train, y_train)
print("Random Forest fitted")
decisionTree.fit(X_train, y_train)
print("Decision Tree fitted")
```

```python
randomForestResult = randomForest.score(X_test, y_test)
decisionTreeResult = decisionTree.score(X_test, y_test)
```

```python
print('Random Forest: {:.2f}'.format(randomForestResult))
print('Decision Tree: {:.2f}'.format(decisionTreeResult))
```
