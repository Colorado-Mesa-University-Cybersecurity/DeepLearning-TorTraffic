### Importing Relevant Libraries

```python
from scipy.io.arff import loadarff
import numpy as np
import pandas as pd
print('relevant libraries uploaded')
```

### Loading arff File

```python
filepath = '../CICDataSet-TOR/CSV/Scenario-A/TimeBasedFeatures-15s-TOR-NonTOR.arff'
raw_data = loadarff(filepath)
```

### Converting into pandas dataframe

```python
df_data = pd.DataFrame(raw_data[0])
```

Fixing that weird character 'b' problem.

```python
df_data['class1'] = df_data['class1'].map(lambda val : 'NONTOR' if val == b'NONTOR' else 'TOR')
```

```python
print(df_data['class1'][0])
```

```python
df_data
```

```python
dep_var = 'class1'
print(df_data[dep_var].value_counts())
```

```python
# Assign the lowest number of samples in a given classification (TOR in this case)
low_count = 3314

# Set a random state for reproducibility
random_state = 1

# Acquire all of the samples that fall into the TOR classification ( WHERE class is 'TOR' )
df1 = df_data.loc[ df_data[dep_var] == 'TOR']

print('Values in df1: ')
print(df1[dep_var].value_counts())

# Acquire all of the samples that fall into the NONTOR class ( WHERE class is 'NONTOR' )
df2 = df_data.loc[ df_data[dep_var] == 'NONTOR']

print('\nValues in df2: ')
print(df2[dep_var].value_counts())
```

## Balancing the dataset

```python
df2 = df2.sample(low_count, random_state=random_state)

print('Values in df1: ')
print(df1[dep_var].value_counts())

print('\nValues in df2: ')
print(df2[dep_var].value_counts())
```

```python
df = pd.concat([df1, df2])

print(df[dep_var].value_counts())
```

```python
df
```

```python
combinedList = list(map(list, zip(df['duration'], df['total_fiat'], df['total_biat'], df['min_fiat'], df['min_biat'], df['max_fiat'], df['max_biat'], df['mean_fiat'], df['mean_biat'], df['flowPktsPerSecond'], df['flowBytesPerSecond'], df['min_flowiat'], df['max_flowiat'], df['mean_flowiat'], df['std_flowiat'], df['min_active'], df['mean_active'], df['max_active'], df['std_active'], df['min_idle'], df['mean_idle'], df['max_idle'], df['std_idle'])))
```

```python
combinedList
```

```python
df['class1'] = df['class1'].map(lambda val : 0 if val == 'NONTOR' else 1)
```

```python
df['class1']
```

```python
df
```

### Class1 column is going to be our target which is 'y' and list of other column (combinedList) is going to be our 'X'

```python
target_value = list()
```

```python
for d in df['class1']:
    target_value.append(d)
```

```python
from sklearn.model_selection import train_test_split
```

```python
random_state = 1
X = combinedList
y = target_value
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=random_state, stratify=y)
```

```python
X_train
```

```python
print("Number of training samples: {}".format(len(X_train)))
print("Number of testing samples: {}".format(len(X_test)))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
randomForest = RandomForestClassifier(max_depth=2, random_state=random_state)
decisionTree = DecisionTreeClassifier(random_state = random_state)
```

```python
randomForest.fit(X_train, y_train)
print("Random Forest fitted")
decisionTree.fit(X_train, y_train)
print("Decision Tree fitted")
```

```python
randomForestResult = randomForest.score(X_test, y_test)
decisionTreeResult = decisionTree.score(X_test, y_test)
```

```python
print('Random Forest: {:.2f}'.format(randomForestResult))
print('Decision Tree: {:.2f}'.format(decisionTreeResult))
```

We have 99% accuracy with both random forest and decision tree machine learning classifier. Now, let us perform some Deep neural network.

# Performing DNN

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.metrics import CategoricalAccuracy, Precision, Recall
from keras.utils import to_categorical
print('Importing relevant libraries for deep learning model')
```

```python
y_train = to_categorical(y_train, num_classes=3)
y_test = to_categorical(y_test, num_classes=3)
```

```python
print('y_train values:')
print(y_train)

print('y_test values:')
print(y_test)
```

```python
accuracy = CategoricalAccuracy() 
prec = Precision()
rec = Recall()
metrics = [accuracy, prec, rec]

model = Sequential()
model.add(Dense(32, input_shape=(23,)))
model.add(Dense(16, activation='relu'))
model.add(Dense(3, activation='softmax'))

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=metrics)
```

```python
print(len(X_train[0]))
```

## converting X-train into numpy arrays into the variable a

```python
a = np.array(X_train)
```

```python
a
```

## Fitting the model

```python
model.fit(a, y=y_train, epochs=20, batch_size=5, verbose=0)
```

## storing the X_test array as numpy arrays into the variable b

```python
b = np.array(X_test)
```

```python
results_dnn = model.evaluate(b, y_test, verbose=0)
acc, prec, rec = results_dnn[1]*100, results_dnn[2], results_dnn[3]
print('Accuracy: {:.2f}%'.format(acc))
print('Precision: {:.2f}'.format(prec))
print('Recall: {:.2f}'.format(rec))
```

The accuracy with deep learning model is 99.77%.
