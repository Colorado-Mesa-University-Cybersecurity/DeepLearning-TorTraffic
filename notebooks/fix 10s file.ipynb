```python
path = '../../tor_dataset/'
scenario = 'Scenario-A/'
file = 'TimeBasedFeatures-15s-TOR-NonTOR.csv'

df1 = pd.read_csv(path + scenario + file)

df1.head()
```

Drop and rename the columns from `df` and then everything should match up! First, let's get rid of the random spaces that show up. These must be artifacts from the FlowMeter program.

```python
df.columns
```

```python
for col in df.columns:
    if col[0] == ' ':
        df[col[1:]] = df[col]
        del df[col]
```

```python
df.columns
```

```python
del df['Source IP']
del df['Source Port']
del df['Destination IP']
del df['Destination Port']
del df['Protocol']
```

I made the code below to be interactive with you! The computer will try to guess which features match up with others and you have to tell it which, with a superior brain, is the correct feature. If there is no correct feature on the page, move to the next page. If there isn't one at all then skip the feature.

```python
from difflib import SequenceMatcher
from ipywidgets import IntSlider, Output
from IPython.display import display, clear_output
out = Output()
display(out)
with out:
    
    #s = SequenceMatcher(None, 'test', 'tdst')
    #s.ratio()

    ratios=[]
    for col in df1.columns:
        ratio = SequenceMatcher(None, 'Active Std', col).ratio()
        ratios.append(ratio)
        #print('{:.2f}\t{}'.format(ratio, col))

    for col in df.columns:
        ratios={}
        for col1 in df1.columns:
            ratios[col1] = SequenceMatcher(None, col, col1).ratio()

        chosen = False
        while chosen is False:
            clear_output()
            print('Candidates for {}'.format(col))
            
            tmp=[]
            i=0
            while i < 5:
                if i >= len(ratios):
                    break

                # Find highest value
                high = max(ratios, key=ratios.get)
                tmp.append(high)

                print('\t{}. {}'.format(i+1, high))

                del ratios[high]
                i+=1
            
            i+=1
            print('\t{}. Next Page'.format(i))
            i+=1
            print('\t{}. Skip Feature'.format(i))

            inp = int(input('Which one matches: ')) - 1
            if inp < (i-2) and inp > -1:
                df[ tmp[inp] ] = df[col]

                if tmp[inp] != col:
                    del df[col]

                chosen = True
            elif inp is (i-1):
                break
            else:
                clear_output()
```

Here is the pattern fyi: 2,3,4,6,2,1,1,1,1,6,1,6,6,6,6,4,2,1,6,6,6,6,4,2,2,6,6,6,6,4,3,3,3,2,1,3,2,2, but make sure to double check just in case!

```python
df.columns
```

We can see that there were some features that we missed! This must be due to a difference in the features present in the dataset. Let's fix those up real quick!

```python
df['std_flowiat'] = df['Flow IAT Std']
del df['Flow IAT Std']

df['std_fiat'] = df['Fwd IAT Std']
del df['Fwd IAT Std']

df['std_biat'] = df['Bwd IAT Std']
del df['Bwd IAT Std']
```

```python
path = '../../tor_dataset/'
scenario = 'Scenario-A/'
file = 'TimeBasedFeatures-10s-TOR-NonTOR.csv'

df.to_csv(path + scenario + file, index=False)
print('File written')
```
